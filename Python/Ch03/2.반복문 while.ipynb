```python
#while
i = 1
while i<=5:
    print('i :', i)
    i += 1
```

```python
# 1부터 10까지 합
total, k = 0, 1
while k<=10:
    total += k
    k += 1

print('1부터 10까지 합 :', total)
```

```python
# 1부터 10까지 짝수합
total, k = 0, 1
while k<=10:
    if k%2==0:
        total += k
    k+=1
print('1부터 10까지 짝수합 :', total)
```

```python
# break
num = 1
while True:
    if num % 5 == 0 and num % 7 == 0:
        break
    num += 1
print('5, 7의 최소공배수 :', num)
```

```python
# continue
num = 0
while num < 10:
    num += 1
    if num % 2 == 0:
        continue
    print(num, end=' ')

```
